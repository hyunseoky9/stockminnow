```python
import matplotlib.pyplot as plt

# Data (already log-transformed!)
reaches = ["Angostura", "Isleta", "San Acacia"]
reach_vals_log = [3, 2.2, 1.2]  # already in log10 scale

wild_site = ["Wild Ne"]
wild_val_log = [2]  # already log10 scale

spring_site = ["Springflow"]
spring_val = [500]   # this one is raw, not logged

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(12, 5))

# --- First subplot: 3 reaches ---
colors_reaches = ["#1b9e77", "#33a02c", "#66c2a5"]
bars1 = axes[0].bar(reaches, reach_vals_log, color=colors_reaches, edgecolor="black")
for bar, val in zip(bars1, reach_vals_log):
    axes[0].text(bar.get_x() + bar.get_width()/2, val + val*0.03,  # dynamic offset (5% of height)
                 f"{val}% are\nage0", ha='center', va='bottom')
axes[0].set_title("Abundance in Reaches")
axes[0].set_ylabel("log10(Abundance)")
axes[0].set_ylim(0, 4)  # keeps it neat

# --- Second subplot: Wild Ne ---
axes[1].bar(wild_site, wild_val_log, color="#d95f02", edgecolor="black")
axes[1].set_title("Wild Effective Population Size")
axes[1].set_ylabel("log10(Wild Ne)")
axes[1].set_ylim(0, 4)

# --- Third subplot: Springflow (linear) ---
axes[2].bar(spring_site, spring_val, color="skyblue", edgecolor="black")
axes[2].set_title("Springflow (KAF)")
axes[2].set_ylabel("Springflow")
axes[2].set_ylim(0, 1100)

# Adjust layout
plt.tight_layout()
plt.show()
```
